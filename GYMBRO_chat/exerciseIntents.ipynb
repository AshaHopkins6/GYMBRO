```python
import csv 
import pandas as pd
```

```python
df =  pd.read_csv('Exercise .csv')
df
```

```python
#main muscle groups 
main_muscle_group = {}
for muscle_group in df['Main Muscle Group'].unique():
    exercises = df.loc[df['Main Muscle Group'] == muscle_group, 'Exercise'].tolist()
    main_muscle_group[muscle_group] = exercises
```

```python
main_muscle_group 
```

```python
#sub muscle groups 
sub_muscle_group = {}
for muscle_group in df['Sub Muscle Group'].unique():
    exercises = df.loc[df['Sub Muscle Group'] == muscle_group, 'Exercise'].tolist()
    sub_muscle_group[muscle_group] = exercises

```

```python
sub_muscle_group
```

```python
intents = []
#List of only the main muscle groups
main_muscle_keys = (list(main_muscle_group.keys()))
# For each muscle group and exercise, create an intent
for muscle_group in main_muscle_keys:
    x = 0
    for x in range(len(main_muscle_group[muscle_group])):
    
        # Create an intent for the muscle group
        intent = {
            "tag": muscle_group,
            "patterns": [f"exercises for {muscle_group}", f"workouts for {muscle_group}", f"I want to target my {muscle_group}", f"I want bigger {muscle_group}", f"I want to grow my {muscle_group}", f"{muscle_group}", f"It is {muscle_group} day!", f"I want a good {muscle_group} pump.", f"Can you recommend a workout for {muscle_group}?", f"What exercises should I do to work on my {muscle_group}?"],
            "responses": [f"Here is an exercise that works your {muscle_group}: {main_muscle_group[muscle_group][x]}", f"Try {main_muscle_group[muscle_group][x]}", f"Add {main_muscle_group[muscle_group][x]} to your routine to target your {muscle_group}"]
        }
        intents.append(intent)

sub_muscle_keys = (list(sub_muscle_group.keys()))
# For each muscle group and exercise, create an intent
for muscle_group in sub_muscle_keys:
    x = 0
    for x in range(len(sub_muscle_group[muscle_group])):
    
        # Create an intent for the muscle group
        intent = {
            "tag": muscle_group,
            "patterns": [f"exercises for {muscle_group}", f"workouts for {muscle_group}", f"I want to target my {muscle_group}", f"I want bigger {muscle_group}", f"I want to grow my {muscle_group}", f"{muscle_group}", f"It is {muscle_group} day!", f"I want a good {muscle_group} pump.", f"Can you recommend a workout for {muscle_group}?", f"What exercises should I do to work on my {muscle_group}?"],
            "responses": [f"Here is an exercise that works your {muscle_group}: {sub_muscle_group[muscle_group][x]}", f"Try {sub_muscle_group[muscle_group][x]}", f"Add {sub_muscle_group[muscle_group][x]} to your routine to target your {muscle_group}"]
        }
        intents.append(intent)


    """
    # For each exercise, create an intent
    for exercise in exercises:
        intent = {
            'tag': exercise,
            'patterns': [f'how to do {exercise}', f'what is {exercise}'],
            'responses': [f'Here is how to do {exercise}: <insert YouTube video link here>']
        }
        intents.append(intent)
"""
```
